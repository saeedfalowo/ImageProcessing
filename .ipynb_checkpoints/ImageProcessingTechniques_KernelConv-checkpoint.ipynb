## 1.1 Digital Image Transformation
### 1.1.1 Filtering
Digital filters can be used to blur and sharpen digital images. Filtering can be performed by:

a.) convolution with specifically designed kernels (filter array) in the spatial domain

b.) masking specific frequency regions in the frequency (Fourier) domain



```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2
import tensorflow as tf
import keras
from Functions import CreateCheckerBoard
#CreateCheckerBoard(2)
```

```python
#img_dir = "Images/Checkerboard(4).png"
#img_dir = "Images/Fruits.jpg"
#img_dir = "Images/Strawberries.jpg"
img_dir = "Images/Lenna.png"
```

```python
# OpenCv load and show img
img = cv2.imread(img_dir)
#img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#cv2.imshow('image',img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
img_arr = np.asarray(img)
plt.imshow(img_arr)#, cmap=plt.get_cmap('gray'))
```

#### Kernels/Filters/Convolution matrix/Mask
These terms are what are used to describe a matrix, a data structure used in image processing to perform tasks such as:

Blurring, Sharpening, Noise removal, Embossing, Edge detection, etc.

Images are N-d arrays, N dimentional arrays, a matrix.

Convolutions are merely matrix multiplications. Image processing is performed by performing a convolution between the kernel/filter matrix and the image matrix. i.e. multiplying the filter matrix with the image matrix.

Performing a convolution using an identity filter/kernel, will output the original image without any changes



1 x n = n & 0 + n = n

Therefore, for multiplication operations, 1 is the identity. While 0 is the identify for addition operation.

```python
def Orignal_v_Filtered(img,filt_img):
    plt.subplot(1,2,1)
    plt.imshow(img)
    plt.title('Original Image')
    
    plt.subplot(1,2,2)
    plt.imshow(filt_img)
    plt.title('Filtered Image')
    
    plt.show()
```

```python
Orignal_v_Filtered(img_arr,img_arr)
```

```python
# Create the first identity kernel
k = np.array(([0,0,0],[0,1,0],[0,0,0]),np.float32)
#k = np.array(np.ones((3,3),np.float32))
print(k)
print(type(k))
```

```python
# Perform the convolutions between the filter matrix and the image array
filt_img = cv2.filter2D(img_arr,-1,k)
Orignal_v_Filtered(img_arr,filt_img)
```

##### EDGE Detection Kernels

```python
# Edge detection kernel 1
k = np.array(([1,0,-1],[0,0,0],[-1,0,1]),np.float32)
print(k)
print(type(k))
filt_img = cv2.filter2D(img_arr,-1,k)
Orignal_v_Filtered(img_arr,filt_img)
```

```python
# Edge detection kernel 2 (Spatial High-pass filter) aka Laplacian Edge Detection
k = np.array(([0,1,0],[1,-4,1],[0,1,0]),np.float32)
print(k)
print(type(k))
filt_img = cv2.filter2D(img_arr,-1,k)
Orignal_v_Filtered(img_arr,filt_img)
```

```python
# Edge detection kernel 3
k = np.array(([-1,-1,-1],[-1,8,-1],[-1,-1,-1]),np.float32)
print(k)
print(type(k))
filt_img = cv2.filter2D(img_arr,-1,k)
Orignal_v_Filtered(img_arr,filt_img)
```

#### High Pass Filters
This can be used to sharpen images

```python
# Image Sharpening kernel
k = np.array(([0,-1,0],[-1,5,-1],[0,-1,0]),np.float32)
print(k)
print(type(k))
filt_img = cv2.filter2D(img_arr,-1,k)
Orignal_v_Filtered(img_arr,filt_img)
```

```python
# Unsharp masking 5x5 kernel
k = np.array(([1,4,6,4,1],[4,16,24,16,4],[6,24,-476,24,6],[4,16,24,16,4],[1,4,6,4,1]),np.float32)
k = (-1/256)*k
print(k)
print(type(k))
filt_img = cv2.filter2D(img_arr,-1,k)
Orignal_v_Filtered(img_arr,filt_img)
```

##### Low-Pass filters
This can be used to blur, denoise, and smoothen images

```python
# Box blur kernel (Spatial low-pass filter)
#k = np.array(([1,1,1],[1,1,1],[1,1,1]),np.float32) # or
#k = np.array(np.ones((3,3),np.float32))
#k = (1/9)*k
# Increase the size of the kernel matrix and number used to divide to see more effects of the filtering
k = np.array(np.ones((7,7),np.float32))
k = (1/45)*k
print(k)
print(type(k))
filt_img = cv2.filter2D(img_arr,-1,k)
Orignal_v_Filtered(img_arr,filt_img)
```

```python
# OpenCv boxFilter blur
#filt_img = cv2.boxFilter(img_arr,-1,(7,7))
filt_img = cv2.boxFilter(img_arr,-1,(20,20))
Orignal_v_Filtered(img_arr,filt_img)
```

```python
# OpenCv blur
#filt_img = cv2.blur(img_arr,(7,7))
filt_img = cv2.blur(img_arr,(20,20))
Orignal_v_Filtered(img_arr,filt_img)
```

```python
# Gaussian blur 3x3 kernel
k = np.array(([1,2,1],[2,4,2],[1,2,1]),np.float32)
k = (1/16)*k
print(k)
print(type(k))
filt_img = cv2.filter2D(img_arr,-1,k)
Orignal_v_Filtered(img_arr,filt_img)
```

```python
# Gaussian blur 5x5 kernel
k = np.array(([1,4,6,4,1],[4,16,24,16,4],[6,24,36,24,6],[4,16,24,16,4],[1,4,6,4,1]),np.float32)
k = (1/256)*k
print(k)
print(type(k))
filt_img = cv2.filter2D(img_arr,-1,k)
Orignal_v_Filtered(img_arr,filt_img)
```

```python
# OpenCv Gaussianblur
#filt_img = cv2.GaussianBlur(img_arr,(3,3),0)
#filt_img = cv2.GaussianBlur(img_arr,(5,5),0)
filt_img = cv2.GaussianBlur(img_arr,(25,25),0)
#filt_img = cv2.blur(img_arr,(20,20))
Orignal_v_Filtered(img_arr,filt_img)
```
